```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
# Load Home Prices
data_frame = pd.read_csv("homeprices.csv")
arr = data_frame[['price']]
print(arr)
```

```python
plt.xlabel("Area (Sq Feet)")
plt.ylabel("Price (US$)")
plt.scatter(data_frame.area, data_frame.price, marker="x", color="red")
```

```python
regression = linear_model.LinearRegression()
regression.fit(data_frame[['area']], data_frame[['price']])
print(regression.coef_)
print(regression.intercept_)
```

```python
d = pd.read_csv("areas.csv")
p = regression.predict(d)
d['price'] = p
```

```python
d.to_csv('predictions.csv')
```
