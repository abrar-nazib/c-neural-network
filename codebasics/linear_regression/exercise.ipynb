```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

## Data Loading and Visualization

```python
training_data = pd.read_csv('canada_per_capita_income.csv')
training_data.head(5)
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.scatter(training_data.year, training_data['per capita income (US$)'], marker='x', color='red')
```

## Training Model

```python
lr_model = linear_model.LinearRegression()
lr_model.fit(training_data[['year']], training_data['per capita income (US$)'])
```

## Prediction visualization with trained model

```python
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.scatter(training_data.year, training_data['per capita income (US$)'], marker='x', color='red')
plt.plot(training_data['year'], lr_model.predict(training_data[['year']]), color='blue')
```
