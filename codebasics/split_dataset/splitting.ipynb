# Splitting Dataset into Test and Train Sets

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Load data

```python
df = pd.read_csv('carprices.csv')
df.head()
```

## Visualize relationships between Independent variables and dependent variables

```python
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Mileage vs Price')
plt.scatter(df[['Mileage']].values, df[['Price']].values)
```

```python
plt.xlabel('Age')
plt.ylabel('Price')
plt.title('Age vs Price')
plt.scatter(df[['Age']].values, df[['Price']].values)
```

So, The dataset seems to be linearly regressable
## Separating Features and Outputs

```python
X = df[['Mileage', 'Age']].values
X
```

```python
Y = df[['Price']].values
Y
```

## Split Dataset into Training Dataset and Test Dataset

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(len(X_train))
print(len(X_test))
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)
lr_model.score(X_test, Y_test)
```
