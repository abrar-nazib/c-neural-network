```python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
train_x = np.array([1000, 2000])
train_y = np.array([300000, 500000])
```

Model representation: $$ f_{w,x}(x)=wx+b $$

```python
# useful functions
def plotData(ax, x, y):
    # Function to scatter plot training data
    ax.scatter(x, y, color="red", marker="x")


def drawLine(ax, x, y):
    # Function to show the regresssion line
    ax.plot(x, y)


def testModel(ax, x, y_train, f_wb):
    # Function to test regression model with the training data
    ax.cla()
    plotData(ax, x, y_train)
    drawLine(ax, x, f_wb)


def feedForward(x, w, b):
    # Function for feed forwarding with specified weight and bias
    f_wb = np.zeros(x.shape[0])
    for i in range(x.shape[0]):
        f_wb[i] = w*x[i]+b
    return f_wb
```

```python
w = 50
b = 100
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.35)

plt.title("House Price")
plt.xlabel("House Size in sq. feet")
plt.ylabel("House Price in $")



f_wb = feedForward(train_x, w, b)
testModel(ax, train_x, train_y, f_wb)


axslider1 = plt.axes([0.1, 0.2, 0.8, 0.05])
slider1 = Slider(axslider1, label="w", valmin=0, valmax = 1000)

plt.show()
```
