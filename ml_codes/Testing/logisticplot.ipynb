```python
"""
This file stores the raw implementation of all ML functions.
Author: Nazib Abrar
"""

def sigmoid_arr(z: np.ndarray):
    """
    Computes the sigmoid of z on an numpy array.

    Args:
        z(ndarray): An array containing values

    Returns:
        g (ndarray): An array containing sigmoid values
    """
    g = 1/(1+np.exp(-z))
    return g

```

```python
z = np.arange(-10,11)

fig,ax = plt.subplots(1,1,figsize=(5,3))
# Plot z vs sigmoid(z)
ax.plot(z, sigmoid_arr(z), c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
plt.plot()
```
