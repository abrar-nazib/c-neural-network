# Line Plots with Seaborn

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
pd.plotting.register_matplotlib_converters() # Custom converters for pandas to work seamlessly with matplotlib
import seaborn as sns
```

```python
# Load Data
spotify_data = pd.read_csv('spotify.csv', index_col='Date', parse_dates=True)
spotify_data.tail()
```

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=spotify_data)
```
